```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
#generating random numbers
n = 1000   
x = np.random.poisson(10, n)    

#create the histogram bins
width = 0.5
histmin = np.floor(min(x)) 
histmax = np.ceil(max(x) + width)  

#plotting the histogram
bins = np.arange(histmin, histmax, width)
plt.hist(x, bins=100, alpha=0.5, edgecolor='cyan')
plt.ylabel("N per bin")
plt.xlabel('x')
plt.savefig('session-10.pdf', bbox_inches='tight', facecolor='white')
```
