```python
import numpy as np
from mpl_toolkits.mplot3d import axes3d  #3D axes with matplotlib
import matplotlib.pyplot as plt
```

```python
fig = plt.figure(figsize=(7,7))   #7x7 image
ax = fig.add_subplot(111, projection='3d')   #single subplot
x = np.linspace(0, np.pi, 200)
y = np.linspace(0, 2*np.pi, 200)
x, y = np.meshgrid(x, y)
z = -np.sin(x) * np.sin(y)

#create a 3D wire frame of the z surface
ax.plot_wireframe(x,y,z, rstride=5, cstride=5, linewidth=0.5, color='0.25')
ax.view_init(20)           #rotate the viewing angle to 20º
ax.set_xlabel('x label')   #add x label
ax.set_ylabel('y label')   #add y label
ax.set_zlabel('z label')   #add z label
plt.savefig('wireframe-surface.png', bbox_inches='tight', facecolor='white')   #save the plot
```

```python
#fancier 3d mesh w/ contour
from matplotlib import cm

fig = plt.figure(figsize=(7,7))   #7x7 image
ax = fig.add_subplot(111, projection='3d')   #single subplot
x = np.linspace(0, np.pi, 200)
y = np.linspace(0, 2*np.pi, 200)
x, y = np.meshgrid(x, y)
z = -np.sin(x) * np.sin(y)

#create a surface plot
surf = ax.plot_surface(x,y,z, rstride=4, cstride=4, cmap=cm.coolwarm, alpha=0.7)

#add a color bar for the surface plot
fig.colorbar(surf, shrink=0.5, label='z label')

#overplot a wireframe on top of the surface
ax.plot_wireframe(x,y,z, rstride=10, cstride=10,linewidth=0.5, color='0.5')

#add the contour plot projection of z projected onto (x,y)
cset = ax.contourf(x,y,z, zdir='z', cmap=cm.coolwarm, offset=-1)

ax.view_init(10)           #rotate the viewing angle to 10º
ax.set_xlabel('x label')   #add x label
ax.set_ylabel('y label')   #add y label

#remove extra z-label on the surface plot
ax.w_zaxis.line.set_lw(0.)
ax.set_zsticks([])

#save the figure
plt.savefig('contour-surface.png', bbox_inches='tight', facecolor='white')
```
